```python
def findMedianSortedArrays(nums1, nums2):
    """
    :type nums1: List[int]
    :type nums2: List[int]
    :rtype: float
    """

    m = len(nums1)
    n = len(nums2)

    if(len(nums2) < len(nums1)):
        nums2, nums1, n, m = nums1, nums2, len(nums1), len(nums2)

    partition1 = m//2
    partition2 = (m+n+1) // 2 - partition1

    # print('\npartition1 - {}'.format(partition1))
    # print('partition2 - {}'.format(partition2))
    start = 0
    med = 0

    while True:

        # import pdb
        # pdb.set_trace()

        if(partition1 == 0):
            leftA = float('-inf')
        else:
            leftA = nums1[partition1-1]

        if(partition1 >= m):
            rightA = float('inf')
        else:
            rightA = nums1[partition1]

        if(partition2 == 0):
            leftB = float('-inf')
        else:
            leftB = nums2[partition2 - 1]

        if(partition2 >= n):
            rightB = float('inf')
        else:
            rightB = nums2[partition2]

        # print('leftA - {}'.format(leftA))
        # print('leftB - {}'.format(leftB))
        # print('rightA - {}'.format(rightA))
        # print('rightB - {}'.format(rightB))

        if (leftA <= rightB) & (leftB <= rightA):
            if((m+n) % 2 == 0):
                med = float(max(leftA, leftB) + min(rightA, rightB)) / 2
            else:
                med = max(leftA, leftB)
            break
        if (leftA > rightB):
            partition1 = (start + partition1) // 2
            partition2 = (m + n + 1) // 2 - partition1
        if (leftB > rightA):
            start = partition1
            partition1 = (partition1 + m) // 2 + (partition1+m) % 2
            partition2 = (m + n + 1) // 2 - partition1


        # print('\npartition1 - {}'.format(partition1))
        # print('partition2 - {}'.format(partition2))
        # c += 1
    return med
```
